```
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

options = webdriver.ChromeOptions()
options.add_argument('-headless')  #invoking chrome in a headless mode(execution of full version of chrome)
options.add_argument('-no-sandbox') #used to permit or restrict modules
options.add_argument('-disable-dev-shm-usage') #shared virtual memory system specific device drivers and programs

#wait tells WebDriver to poll the DOM(Document Object Model) 
#for a certain amount of time when trying to find any element not immediately available
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.common.by import By
import pandas as pd
import time

#Capabilities are options used to customize and configure a ChromeDriver session
driver = webdriver.Chrome('chromedriver',options=options)
reviews_link = 'https://www.imdb.com/title/tt7286456/reviews?ref_=tt_urv'
arr_T= []
arr_R = []
driver.get(reviews_link)

for i in range(20):

  driver.find_element_by_class_name("ipl-load-more__button").click() #using class name name to find the appropritae elements
  time.sleep(5)
  Tit = driver.find_elements(By.CLASS_NAME, "title") #finding elements that are used as title of the review
  Rev = driver.find_elements(By.CLASS_NAME, "text") #finsing elements that is the text of the reviews

for ele, sub_ele in zip(Tit, Rev):

      arr_T.append((ele.text).replace('\n',''))
      arr_R.append(sub_ele.text)

#zip() is used to iterate elements from two or more iterables
df = pd.DataFrame(list(zip(arr_T, arr_R)), columns =['Title of the Review', 'The Review'])
print("Length of data frame: {0}".format(len(df)))

#printing the data in the dataframe

df
```

```
import re
df['Reviews after removing noise: special characters and punctuation']=df['The Review'].str.replace('[^\w\s]','')
df['Reviews after removing noise: special characters and punctuation'] = df['Reviews after removing noise: special characters and punctuation'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', character) for character in x ))
```

```
df['Reviews after removing digits'] = df['Reviews after removing noise: special characters and punctuation'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
```

```
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['Reviews after removing StopWords'] = df['Reviews after removing digits'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```
df['Reviews after converting into LowerCase'] = df['Reviews after removing StopWords'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```
from textblob import Word
import nltk
nltk.download('wordnet')

df['Reviews after preprocessing'] = df['Reviews after converting into LowerCase'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```

```
#POS tagging

from nltk.tokenize import word_tokenize
pos = []
for sent in df['Reviews after preprocessing']:
  txt = word_tokenize(sent)
  pos.append(nltk.pos_tag(txt))
pos
```

```
Noun = []
Verb = []
Adjective = []
Adverb = []
Cordinating_Conjunction = []
Subordinating_Conjuction = []
Interjection = []
Personal_Pronoun = []
Predeterminer = []
Determiner = []
```

```
for value in pos:
  
  Noun_Count = 0
  Verb_Count = 0
  Adjective_Count = 0
  Adverb_Count = 0
  Cordinating_Conjunction_Count = 0
  Subordinating_Conjuction_Count = 0
  Interjection_Count = 0
  Personal_Pronoun_Count = 0
  predeterminer_Count = 0
  Determiner_Count = 0

  #using for loop to iterate and store
  for word,tag in value:
    
    if tag == 'NN':
      Noun_Count = Noun_Count + 1
    elif tag == 'VR':
      Verb_Count = Verb_Count + 1
    elif tag == 'JJ':
      Adjective_Count = Adjective_Count + 1
    elif tag == 'RB':
      Adverb_Count = Adverb_Count + 1
    elif tag == 'CC':
      Cordinating_Conjunction_Count = Cordinating_Conjunction_Count + 1
    elif tag == 'IN':
      Subordinating_Conjuction_Count = Subordinating_Conjuction_Count + 1
    elif tag == 'UH':
      Interjection_Count = Interjection_Count + 1
    elif tag == 'PRP':
      Personal_Pronoun_Count = Personal_Pronoun_Count + 1
    elif tag == 'PDT':
      predeterminer_Count = predeterminer_Count + 1
    elif tag == 'DT':
      Determiner_Count = Determiner_Count + 1
   
  #append adds single item to existing item, helps in modifying original list
  Noun.append(Noun_Count)
  Verb.append(Verb_Count)
  Adjective.append(Adjective_Count)
  Adverb.append(Adverb_Count)
  CordinatingConjunction.append(Cordinating_Conjunction_Count)
  SubordinatingConjuction.append(Subordinating_Conjuction_Count)
  Interjection.append(Interjection_Count)
  PersonalPronoun.append(Personal_Pronoun_Count)
  predeterminer.append(predeterminer_Count)
  Determiner.append(Determiner_Count)
```

```
#storing in  df

df['Noun Count'] = Noun
df['Verbs Count'] = Verb
df['Adjectives Count'] = Adjective
df['Adverbs Count'] = Adverb
df['Cordinating Conjunctions Count'] = CordinatingConjunction
df['Subordinating Conjuctions Count'] = SubordinatingConjuction
df['Interjections Count'] = Interjection
df['Personal Pronouns Count'] = PersonalPronoun
df['Predeterminers Count'] = predeterminer
df['Determiners Count'] = Determiner

#printing df
df
```
