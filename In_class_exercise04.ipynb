```python
#n-grams

import collections
import re
import sys
import time

def tokenize(string):
    return re.findall(r'\w+', string.lower())

def n_grams(lines, min=1, max=3):
    leng = range(min, max + 1)
    ngrams = {length: collections.Counter() for length in leng}
    queue = collections.deque(maxlen=max)

    def add_queue():
        current = tuple(queue)
        for length in leng:
            if len(current) >= length:
                ngrams[length][current[:length]] += 1

    for line in lines:
        for word in tokenize(line):
            queue.append(word)
            if len(queue) >= max:
                add_queue()

    while len(queue) > min:
        queue.popleft()
        add_queue()

    return ngrams

def pmf(ngrams, num=10):
    for n in sorted(ngrams):
        print('*****Top {} {}-grams *****'.format(num, n))
        for g, count in ngrams[n].most_common(num):
            print('{0}: {1}'.format(' '.join(g), count))
        print('')

if __name__ == '__main__':
    if len(sys.argv) < 2:
        print('Usage: python ngrams.py filename')
        sys.exit(1)

    start_time = time.time()
    with open('legalcase.txt', 'r') as v:
        ngrams = n_grams(v)
    pmf(ngrams)
  
```
